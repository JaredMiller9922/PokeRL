# **Prompting Itself on What to do Next**

````python
import os
import sys
import json
from PIL import Image
from pathlib import Path
from collections import deque
import matplotlib.pyplot as plt

# Path Setup
base_path = "/content/drive/MyDrive/Personal/School/02 - UofU Grad/CS 6960 - Multimodal Agents/Class Project/PokemonRedExperiments/PokemonRedExperiments-master"
v2_path = os.path.join(base_path, "v2")

os.chdir(v2_path)
sys.path.append(v2_path)

from red_gym_env_v2 import RedGymEnv

# Configuration
env_config = {
    'headless': True,
    'save_final_state': False,
    'early_stop': False,
    'action_freq': 24,
    'init_state': '../has_pokedex_nballs.state',
    'max_steps': 2048,
    'print_rewards': False,
    'save_video': True,
    'fast_video': False,
    'session_path': Path('session'),
    'gb_path': '../PokemonRed.gb',
    'debug': False,
    'sim_frame_dist': 2_000_000.0
}

# Initialization
env = RedGymEnv(env_config)
obs, info = env.reset()
reward = 0

action_map = {
    "down": 0,
    "left": 1,
    "right": 2,
    "up": 3,
    "a": 4,
    "b": 5,
    "start": 6
}

frame_buffer = deque(maxlen=5)
recent_actions = deque(maxlen=15)

step = 0

# Starting sequence to get out of Prof. Oak's house
for move, count in [("down", 15), ("left", 5)]:
    for _ in range(count):
        screen = obs
        screen = next(val for val in screen.values() if hasattr(val, 'shape') and len(val.shape) >= 2)
        if screen.shape[0] == 3:
            screen = screen.transpose(1, 2, 0)

        frame_buffer.append(Image.fromarray(screen))

        obs, reward, term, trunc, info = env.step(action_map[move])
        recent_actions.append(move)

        print(f"Step {step}: {move.upper()} | Map: {info.get('map_id')}")
        step += 1

        if term or trunc:
            break

# chat = client.chats.create(model="gemini-3-flash-preview")
chat = client.chats.create(model="gemini-3.1-flash-lite-preview")

# Main Autonomous Loop
while step < 2048:
    current_obs = obs
    current_obs = next(val for val in current_obs.values() if hasattr(val, 'shape') and len(val.shape) >= 2)
    if current_obs.shape[0] == 3:
        current_obs = current_obs.transpose(1, 2, 0)

    img = Image.fromarray(current_obs)
    frame_buffer.append(img)

    fig, axes = plt.subplots(1, len(frame_buffer), figsize=(12, 3))
    axes = [axes] if len(frame_buffer) == 1 else axes

    for i, frame in enumerate(frame_buffer):
        axes[i].imshow(frame)
        axes[i].axis('off')
        axes[i].set_title(f"Frame {i}")

    plt.show()

    current_x = info.get('x', 0)
    current_y = info.get('y', 0)
    map_id = info.get('map_id', 'Unknown')
    goal_text = ""
    last_move = recent_actions[-1] if recent_actions else "none"

    prompt = f"""
    You are an expert navigator playing Pokémon Red. Your objective is to reach Viridian City by traveling NORTH through Route 1.

    Current Coordinates: (X: {current_x}, Y: {current_y})
    Last Move Executed: {last_move}
    Goal: {goal_text}

    Obstacle Avoidance & Navigation Rules:
    1. COLLISION DETECTION: If Current Coordinates equal Previous Coordinates, your Last Move hit a wall or obstacle. You MUST immediately change to an orthogonal direction (e.g., if blocked moving 'up', output 'left' or 'right' for the next moves).
    2. VISUAL PATHFINDING: Analyze the immediate tiles in your frame. Explicitly route around ledges (unless jumping down), fences, trees, and NPCs.
    3. BYPASS MANEUVER: If your primary direction (North) is visually blocked, step horizontally (East/West) to clear the obstacle before resuming North.
    4. ANTI-LOOPING: Do not repeat a sequence of moves that previously resulted in no coordinate change.
    5. INTERACTION: Use 'a' to clear unexpected dialogue boxes or if forced to interact with a blocking object.
    6. LEDGE DETECTION: Look closely for thick, shadowy horizontal dividing lines immediately above the player. These are ledges and cannot be walked over from the bottom.

    Response Format (Strict JSON):
    {{
      "analysis": "State if you are currently stuck, identify visible obstacles, and determine the clear path.",
      "objective": "State the immediate micro-step to bypass the obstacle.",
      "moves": ["list", "of", "moves", "here"]
    }}
    """.strip()


    # prompt = f"""
    # You are an expert navigator playing Pokémon Red. Exit Pallet Town and travel NORTH through Route 1 to reach Viridian City.

    # Current Coordinates: (X: {current_x}, Y: {current_y})
    # Last Move Executed: {last_move}
    # Goal: {goal_text}

    # Rules for Move Sequences:
    # 1. Do not return 10 identical moves unless you are in a long, empty hallway.
    # 2. Use a mix of directions to navigate around obstacles visible in the frames.
    # 3. Use 'a' if you are near an NPC, sign, door, or dialogue box.
    # 4. Focus on making progress north toward Viridian City.

    # Response Format (Strict JSON):
    # {{
    #   "analysis": "...",
    #   "objective": "...",
    #   "moves": ["down", "down", "right", "down", "a", "down", "down", "left", "down", "down"]
    # }}
    # """.strip()

    print(f"\n{'='*20} Prompting Gemini {'='*20}")

    response = chat.send_message(list(frame_buffer) + [prompt])
    clean_json = response.text.replace("```json", "").replace("```", "").strip()
    data = json.loads(clean_json)

    print(f"ANALYSIS: {data.get('analysis')}")
    print(f"OBJECTIVE: {data.get('objective')}")
    action_sequence = data.get('moves', [])

    print(f"--- Executing Plan: {action_sequence} ---")

    for action_text in action_sequence:
        action_key = action_text.lower().strip()

        obs, reward, term, trunc, info = env.step(action_map[action_key])
        recent_actions.append(action_key)

        print(f"Step {step}: {action_key.upper()} | Map: {info.get('map_id')} | Reward: {reward}")
        step += 1

        if info.get('map_id', map_id) != map_id or info.get('is_in_battle', False):
            print("!!! State changed. Re-planning objective...")
            break

        if term or trunc:
            print("Episode ended. Resetting...")
            obs, info = env.reset()
            frame_buffer.clear()
            recent_actions.clear()

            for _ in range(10):
                obs, reward, term, trunc, info = env.step(action_map["down"])

            chat = client.chats.create(model="gemini-3.1-flash-lite-preview")
            break

env.close()
````
